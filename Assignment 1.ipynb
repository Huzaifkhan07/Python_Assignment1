6. Write a code to perform basic string compression using the counts of repeated characters

```python
string = "tttttuuuutorrrriaaaaalllllll"

strli =[]
count = 1
pre = string[0]

for i in range(1, len(string)):
  pointer = string[i]
  if pre == pointer:
    count+=1
  else:
    strli.append(pre)
    strli.append(str(count))
    pre = pointer
    count = 1

strli.append(pre)
strli.append(str(count))
compressed  = "".join(strli)

if len(compressed) < len(string):
  print(compressed)
else:
  print(string)

```

14. Write a code to reverse a list in-place without using any built-in reverse functions

```python
def revlist(li):
  left = 0
  right = len(li)-1

  while left < right:
    li[left],li[right] = li[right],li[left]
    left+=1
    right-=1

fruits = ['apple', 'banana', 'cherry', 'mango', 'orange','papaya']
revlist(fruits)
fruits
```

20. Write a code to shuffle a given list randomly without using any built-in shuffle functions

```python
import random

lis1 = ["apple", "banana", "mango", "orange", "pineapple"]

# Applying Fisher-Yates shuffle algorithm
def shuffle(lst):
  n = len(lst)
  for i in range(n):
    j = random.randint(i, n-1)
    lst[i], lst[j] = lst[j], lst[i]
  return lst

print(shuffle(lis1))
```

21.  Write a code that takes two tuples as input and returns a new tuple containing elements that are
common to both input tuples

```python
def input_to_tuple(prompt):
    user_input = input(prompt)
    return tuple(user_input.split(','))

def common_elements(tup1, tup2):
    common = ()
    for i in tup1:
        if i in tup2:
            common += (i,)
    return common

# Taking two tuples as input from the user
tup1 = input_to_tuple("Enter the first tuple (comma-separated values): ")
tup2 = input_to_tuple("Enter the second tuple (comma-separated values): ")

# Finding and printing the common elements
result = common_elements(tup1, tup2)
print(f"The common elements are: {result}")
```

30. Develop a code that prompts the user to input two sets of strings. Then, print the symmetric difference of
these two sets

```python
# symmetric difference : it includes elements that are unique to each set.
def makeSymDiff(inp1, inp2):
    set1 = set(map(str.strip, inp1.split(',')))
    set2 = set(map(str.strip, inp2.split(',')))
    symDiff = set1.symmetric_difference(set2)
    return symDiff

inp1 = input("Enter the first set of strings (comma-separated values): ")
inp2 = input("Enter the second set of strings (comma-separated values): ")

print(f"The symmetric difference of the given sets is : {makeSymDiff(inp1, inp2)}")
```

32.  Write a code that takes two dictionaries as input and merges them into a single dictionary. If there are
common keys, the values should be added together

```python
def merging_dicts(dict1, dict2):
    merged_dict = dict1.copy()
    for key, value in dict2.items():
        if key in merged_dict:
            merged_dict[key] += value
        else:
            merged_dict[key] = value
    return merged_dict


dict1 = {'apple': 3, 'banana': 2, 'cherry': 5}
dict2 = {'apple': 2, 'banana': 1, 'mango': 4}

print("Merged Dictionary:", merging_dicts(dict1, dict2))
```

33.  Write a code to access a value in a nested dictionary. The function should take the dictionary and a list of
keys as input, and return the corresponding value. If any of the keys do not exist in the dictionary, the
function should return None

```python
def ReturnValue(myfamily, children):
    val = {}
    for key in children:
        if key in myfamily:
            val[key] = myfamily[key].get("name")
        else:
            val[key] = None
    return val

myfamily = {
    "child1": {"name": "Vickey"},
    "child2": {"name": "John"},
    "child3": {"name": "Dany"}
}
children = ["child1", "child2", "child3", "child4"]

result = ReturnValue(myfamily, children)
print(result)
```

35. Write a code that inverts a dictionary, swapping keys and values. Ensure that the inverted dictionary
correctly handles cases where multiple keys have the same value by storing the keys as a list in the
inverted dictionary.

```python
def invDict(fruits):
  invertedDict= {}
  for key, value in fruits.items():
    if value not in invertedDict:
      invertedDict[value] = [key]
    else:
      invertedDict[value].append(key)
  return invertedDict


fruits = {'apple': 5, 'banana': 3, 'cherry': 2, 'mango': 5, 'orange': 2, 'lichi': 8, 'papaya': 2}
result = invDict(fruits)
print(result)
```
